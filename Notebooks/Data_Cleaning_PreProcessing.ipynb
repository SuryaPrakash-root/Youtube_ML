```python
# Load the Data From CSV File
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
video_data = pd.read_csv('../Data/USvideos.csv')
video_data.head()
```

```python
video_data.describe()
```

```python
video_data.info()
```

```python
#Exploratory Data Analysis (EDA)
video_data.duplicated().sum()
video_data = video_data.drop_duplicates()
```

```python
video_data.isna().sum()
```

```python
Fill_Na = ['likes', 'comments', 'watch_time_minutes']

video_data[Fill_Na] = video_data[Fill_Na].fillna(video_data[Fill_Na].median())

video_data.isna().sum()
```

```python
video_data.info()
video_data.describe()
```

```python
video_data
```

```python
video_data['date'] = pd.to_datetime(video_data['date'])
video_data['trending_date'] = pd.to_datetime(video_data['date'].dt.date)
video_data['trending_time'] = video_data['date'].dt.strftime('%H:%M:%S')
video_data['likes'] = video_data['likes'].astype(np.int64)
video_data['comments'] = video_data['comments'].astype(np.int64)
video_data['watch_time_minutes'] = video_data['watch_time_minutes'].astype(np.int64)
video_data['video_length_minutes'] = video_data['video_length_minutes'].round(2).astype(np.float64)
video_data['ad_revenue_usd'] = video_data['ad_revenue_usd'].round(2).astype(np.float64)
video_data
```

```python
video_data.drop(['date', 'video_id'], axis=1, inplace=True)
video_data
```

```python
# Feature Engineering
video_data['engagement_rate'] = (video_data['likes']+ video_data['comments'] / video_data['views']* 100).round(2)
video_data
```

```python
#Plotting the Distribution of Engagement Rate
numeruic_features = ['views', 'likes', 'comments', 'watch_time_minutes', 'video_length_minutes', 'ad_revenue_usd', 'engagement_rate']
plt.figure()
plt.boxplot(video_data[numeruic_features])
plt.title(f'Boxplot of {numeruic_features}')
plt.ylabel('numeric_features')
```

```python
# Z-score Normalization
from scipy.stats import zscore

z_score = zscore(video_data[numeruic_features])
Outliers_mask = (np.abs(z_score) > 3).any(axis=1)
Outliers_data = video_data[Outliers_mask]
Outliers_data.shape

# Removing Outliers
video_data_cleaned = video_data[~Outliers_mask]
video_data_cleaned.shape
video_data.describe()
```

```python
# Interquartile Range (IQR) Method
Q1 = video_data[numeruic_features].quantile((0.25))
Q3 = video_data[numeruic_features].quantile((0.75))
IQR = Q3 - Q1
lower_bond = Q1 - 1.5 * IQR
upper_bond = Q3 + 1.5 * IQR
outliers = ((video_data[numeruic_features] < lower_bond) | (video_data[numeruic_features] > upper_bond)).any(axis=1)
outliers_data = video_data[outliers]
outliers_data.shape
```

```python
# Plotting
plt.scatter((Outliers_data['views']), (Outliers_data['engagement_rate']), color='red', label='Outliers')
plt.xlabel('Views')
plt.ylabel('Engagement Rate')
plt.title('Outliers in Views vs Engagement Rate')
plt.legend()
```

```python
# # Encoding Categorical Variables
from sklearn.preprocessing import OneHotEncoder

categorical_cols = ['category', 'device', 'country']

encoding = OneHotEncoder(sparse_output=False, drop='first')
encoded_data = encoding.fit_transform(video_data[categorical_cols])
encoded_df = pd.DataFrame(encoded_data, columns=encoding.get_feature_names_out(categorical_cols))
video_data = pd.concat([video_data.drop(categorical_cols, axis=1), encoded_df], axis=1)
video_data


```
